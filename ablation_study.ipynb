# Speech Commands

```python
import pandas as pd
import numpy as np
import os
from lib.ablationStudyUtils import loss_ablation, format_loss

root_path = './result/speech-commands/CoNMix/ablation_study'
accu_10dB = pd.read_csv(os.path.join(root_path, 'accu-10dB-SC.csv'), sep=',')
accu_10dB.head()
```

```python
x = np.arange(21)
loss_ablation(
    x=x, 
    title='CoNMix Prediction Accuracy (%) under 10dB Running Tap Background Noise \n in Speech Commands',
    lines=format_loss(
        origin=accu_10dB['RT-10dB-org'], updated=accu_10dB['RT-10dB-upd'], no_pl=accu_10dB['RT-10dB-no_pl'],
        no_cst=accu_10dB['RT-10dB-no_cst']
    )
)
```

```python
loss_ablation(
    x=x,
    title='CoNMix Prediction Accuracy (%) under 10dB Exercise Bike Backgournd Noise \n in Speech Commands',
    lines=format_loss(
        origin=accu_10dB['EB-10dB-org'], updated=accu_10dB['EB-10dB-upd'], no_pl=accu_10dB['EB-10dB-no_pl'],
        no_cst=accu_10dB['EB-10dB-no_cst']
    )
)
```

```python
loss_ablation(
    x=x,
    lines=format_loss(
        origin=accu_10dB['DD-10dB-org'], updated=accu_10dB['DD-10dB-upd'], no_pl=accu_10dB['DD-10dB-no_pl'],
        no_cst=accu_10dB['DD-10dB-no_cst']
    ),
    title='CoNMix Prediction Accuracy (%) under 10dB Doing the Dishes Backgournd \n Noise in Speech Commands'
)
```

```python
lines = {
    'running tap': {
        'data': accu_10dB['RT-10dB-no_nm'].loc[:20],
        'linestyle': '-',
        'color': 'b'
    },
    'exercise bike': {
        'data': accu_10dB['EB-10dB-no_nm'].loc[:20],
        'linestyle': '-',
        'color': 'g'
    },
    'doing the dishes': {
        'data': accu_10dB['DD-10dB-no_nm'].loc[:20],
        'linestyle': '-',
        'color': 'r'
    }
}
loss_ablation(
    x=x,
    title='CoNMix Prediction Accuracy (%), which excludes the Nuclear-norm Loss, \n under 10dB Background Noise in Speech Commands',
    lines=lines
)
```
