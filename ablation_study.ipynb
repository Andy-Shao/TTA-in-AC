# Speech Commands

```python
import pandas as pd
import numpy as np
import os
from lib.ablationStudyUtils import loss_ablation, format_loss

root_path = './result/speech-commands/CoNMix/ablation_study'
accu_10dB = pd.read_csv(os.path.join(root_path, 'accu-10dB-SC.csv'), sep=',')
accu_10dB.head()
```

```python
loss_ablation(
    x=x,
    lines=format_loss(
        origin=accu_10dB['DD-10dB-org'], updated=accu_10dB['DD-10dB-upd'], no_pl=accu_10dB['DD-10dB-no_pl'],
        no_cst=accu_10dB['DD-10dB-no_cst']
    ),
    title='CoNMix Prediction Accuracy (%) under 10dB Doing the Dishes Backgournd \n Noise in Speech Commands'
)
```
