```python
import torchvision.models as mds
restNet18 = mds.resnet18()
restNet18
```

```python
import torch.nn as nn

class MyRestNet18(nn.Module):
    def __init__(self):
        super(MyRestNet18, self).__init__()
        self.restNet = mds.resnet18()
        fc = nn.Linear(in_features=512, out_features=10, bias=True)
        conv1 = nn.Conv2d(in_channels=1, out_channels=64, kernel_size=7, stride=2, padding=3, bias=False)
        self.restNet.fc = fc
        self.restNet.conv1 = conv1

    def forward(self, x):
        return self.restNet(x)
    

myRn18 = MyRestNet18()
myRn18
```

```python
import torch 
inputs = torch.ones(32, 1, 224, 224)
outpus = myRn18(inputs)
inputs.shape, outpus.shape
```
